```python
def fPreprocesamiento(df, lista_cctv):
    if "id" in df.columns:
        df_cctv = df[df["id"].isin(lista_cctv)].copy()
    elif "idelem" in df.columns:
        df_cctv = df[df["idelem"].isin(lista_cctv)].copy()
        df_cctv = df_cctv.rename(columns={'idelem': 'id'})
    else:
        return None # Retornar un DataFrame vacÃ­o para evitar errores posteriores

    if not df_cctv.empty:
        # La intensidad esta expresada en vehiculos por hora por lo que
        df_cctv.loc[:, "intensidad"] = df_cctv["intensidad"].apply(lambda x: x/4)
        # Paso la columna fecha a datetime
        df_cctv.loc[:,'fecha_hora'] = pd.to_datetime(df_cctv['fecha'], format='%Y-%m-%d %H:%M:%S')
        # Sumo de cada punto de recogida de datos la intensidad de vehiculos que hay por horas pasando de ir de 15 min en 15 a ir de hora en hora
        df_hora = df_cctv.groupby(['id', pd.Grouper(key='fecha_hora', freq='H')])['intensidad'].sum().reset_index()
        # Creo las columnas aÃ±o mes dia y hora para compatibilizarlo con el data de datos de calidad de aire
        df_hora['aÃ±o'] = df_hora['fecha_hora'].dt.year
        df_hora['mes'] = df_hora['fecha_hora'].dt.month
        df_hora['dia'] = df_hora['fecha_hora'].dt.day
        df_hora['hora'] = df_hora['fecha_hora'].dt.hour
        df_hora = df_hora.drop(columns="fecha_hora", axis=1, errors='ignore') # Especificar axis=1 y manejar errores si no existe
        return df_hora
    else:
        return None
```

```python
def fLecturacsv(ruta, lista):
    try:
        df = pd.read_csv(ruta, sep=";")
        df_final = fPreprocesamiento(df, lista)
        return df_final
    except Exception as e:
        print(f"El archivo {ruta} no se pudo leer: {e}")
        return None  
```

```python
lista_cctv = [4297, 1018, 5783, 6117, 4027,3891, 4754, 5064, 4839, 4124, 5365, 11038, 4436, 5772, 3910, 7017,10184, 3787, 6610, 6333, 6342]
mes = ["01","02","03","04","05","06","07","08","09","10","11","12"]
df_final = pd.DataFrame()
for aÃ±o in range(2021, 2025):
    for i in range(12):
        archivo = mes[i]+"-"+str(aÃ±o)+".csv"
        ruta = os.path.join(str(aÃ±o), archivo)
        df = fLecturacsv(ruta, lista_cctv)
        df_final = pd.concat([df_final,df])
```
