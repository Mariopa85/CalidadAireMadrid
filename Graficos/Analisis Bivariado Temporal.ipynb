Al tratarse de un problema con temporalidad primero mirare como reaccionan las targets con las variables temporales

```python
col_temporales = ['Año', 'Mes', 'Dia', 'Hora']
col_target = ["Dióxido de Azufre", "Dióxido de Nitrógeno", "Partículas<2.5", "Partículas<10", "Ozono"]
for target in col_target:
    fig, ax = plt.subplots(nrows=len(col_temporales), ncols=1, figsize=(15, 7 * len(col_temporales)))
    fig.suptitle(f'Análisis de {target} vs variables temporales')

    for i, col in enumerate(col_temporales):
        sns.barplot(x=col, y=target, data=df, ax=ax[i])
        ax[i].set_xlabel(col)
        ax[i].set_ylabel(target)
        ax[i].set_title(f'Comparación de {col} vs {target}')
        
    plt.tight_layout(rect=[0, 0.03, 1, 0.98])
    plt.show()
```

```python
for temp in col_temporales:
    plt.figure(figsize=(15,10))
    sns.barplot(x=temp, y="Intensidad", data=df)
    plt.show()
```

```python
col_target = ['Dióxido de Azufre', 'Dióxido de Nitrógeno', 'Partículas<2.5', 'Partículas<10', 'Ozono','Estación']
col_temporales = ['Año', 'Mes', 'Dia', 'Hora']
col_a_analizar = [col for col in df.columns if col not in col_target and col not in col_temporales]

for target in col_a_analizar:
    fig, ax = plt.subplots(nrows=len(col_temporales), ncols=1, figsize=(15, 7 * len(col_temporales)))
    fig.suptitle(f'Análisis de {target} vs variables temporales')

    for i, col in enumerate(col_temporales):
        sns.barplot(x=col, y=target, data=df, ax=ax[i])
        ax[i].set_xlabel(col)
        ax[i].set_ylabel(target)
        ax[i].set_title(f'Comparación de {col} vs {target}')

    plt.tight_layout(rect=[0, 0.03, 1, 0.98])
    plt.show()
```
