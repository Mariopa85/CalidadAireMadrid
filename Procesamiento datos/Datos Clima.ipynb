```python
import pandas as pd
```

```python
df = pd.read_csv("Clima20132016.csv")
```

```python
df = df.drop(columns="ALLSKY_SFC_SW_DWN")
```

```python
df.head()
```

```python
dic = {"T2M":"Temperatura", "T2MDEW":"PuntoRocio", "T2MWET":"TempHumeda", "RH2M":"HumedadRelativa", "PRECTOTCORR":"Precipitacion", "PS":"Presion", "WS2M":"VelViento", "WD2M":"DirViento"}
```

```python
df = df.rename(columns=dic)
df.head()
```

```python
#Los datos de trafico solo llegan desde Julio de 2013 por lo que debo eliminar los primeros 6 meses de registros
año = 2013
meses = [1, 2, 3, 4, 5, 6]
filtro = (df["YEAR"]==año) & (df["MO"].isin(meses))
df_filtrado = df[~filtro].copy()
```
