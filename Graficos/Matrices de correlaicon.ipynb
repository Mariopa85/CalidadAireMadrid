```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("datos_finales.csv")
```

```python
matriz_correlacion = df.corr()
```

```python
plt.figure(figsize=(15, 12))
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm', fmt=".2f")
plt.show()
```

```python
col_target = ["Dixido de Nitrgeno","Partculas<2.5","Ozono"]
```

```python
correlation_matrix = df.corr()

# Seleccionar las correlaciones de las columnas con las variables en col_target
correlation_with_target = correlation_matrix[col_target].drop(col_target, axis=0)

# Visualizar la matriz de correlacin
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_with_target, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlacin con Variables Target')
plt.ylabel('Otras Variables')
plt.xlabel('Variables Target')
plt.show()
```
