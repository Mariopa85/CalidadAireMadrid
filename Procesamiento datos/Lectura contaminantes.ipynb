```python
import pandas as pd
import os
```

Lo primero que hare es un enlacce entre estaciones y sus respectivos nombres

```python
def fVal():
    contador = 1
    lista = []
    for i in range(24):
        if contador < 10:
            hora = str(contador)
            col = "V0"+hora
            lista.append(col)
        else:
            hora = str(contador)
            col = "V"+hora
            lista.append(col)
        contador += 1
    return lista
```

```python
def fPreprocesamiento(df):
    df_final = df.drop(columns=fVal())
    df_final = df_final.drop(columns=["PROVINCIA","MUNICIPIO","PUNTO_MUESTREO"])
    if df_final.isna().any().any():
        df_final = df_final.fillna(df_final.drop(columns=["ESTACION","MAGNITUD","ANO","MES","DIA"]).mean(axis=1))
    return df_final
```

```python
def fLecturacsv(ruta):
    df = pd.read_csv(ruta, sep=";")
    df_final = fPreprocesamiento(df)
    return df_final
```

```python
mes = ("ene","feb","mar","abr","may","jun","jul","ago","sep","oct","nov","dic")
año = 2013
df_final = pd.DataFrame()
for year in range(12):
    for i in range(12):
        archivo = mes[i]+"_mo"+str(año)[-2:]+".csv"
        ruta = os.path.join(str(año), archivo)
        df = fLecturacsv(ruta)
        df_final = pd.concat([df_final,df])
    año+=1
```
