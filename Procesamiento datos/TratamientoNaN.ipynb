```python
estacion_39 = df[df["Estación"] == 39].copy()
estacion_39 = estacion_39.reset_index(drop=True)
estacion_50 = df[df["Estación"] == 50].copy()
estacion_50 = estacion_50.reset_index(drop=True)
estacion_60 = df[df["Estación"] == 60].copy()
estacion_60 = estacion_60.reset_index(drop=True)
```

```python
indices_no_na_50 = estacion_50[estacion_50["Intensidad"].notna()].index
indices_no_na_60 = estacion_60[estacion_60["Intensidad"].notna()].index
indices_na_39 =  estacion_39.index[pd.isna(estacion_39['Intensidad'])].tolist()
```

```python
indices_no_na = list(set(indices_no_na_50) & set(indices_no_na_60) & set(indices_na_39)) 
#Creo una lista con los indices de valores no na de las listas 50 y 60 y una de los valores na de la 39 ya que son lo que debo sustituir
```

```python
len(indices_no_na)
```

```python
estacion_39.loc[indices_no_na, "Intensidad"] = (estacion_50.loc[indices_no_na, "Intensidad"]/2)-(estacion_60.loc[indices_no_na, "Intensidad"]*3)
```

```python
estacion_39.groupby(["Estación"])["Intensidad"].apply(lambda x: x.isna().sum())
```

```python
df.loc[df["Estación"] == 39, "Intensidad"] = estacion_39["Intensidad"].values
```
