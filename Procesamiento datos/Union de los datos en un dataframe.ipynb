```python
df_clima = pd.read_csv("datos_clima.csv")
df_clima
```

```python
#Me he dado cuenta que los datos del clima van de 0 a 23 a 23 en vez de 1 a 24, para arreglar esto quitare la primera fila subiendo la columna una posicion del df y cambiare los 0 por 24
df_clima[["Hora","Temperatura","PuntoRocio","TempHumeda","HumedadRelativa", "Precipitacion","Presion","VelViento","DirViento"]] = df_clima[["Hora","Temperatura","PuntoRocio","TempHumeda","HumedadRelativa", "Precipitacion","Presion","VelViento","DirViento"]].shift(-1)
df_clima["Hora"] = df_clima["Hora"].replace(0, 24)
```

```python
df_clima
```

```python
#Para completar la fila de NaN que queda en el ultimo registro de datos climatologicos la relleno con los mismos valores que fila anterior 
df_clima = df_clima.ffill()
#Arreglo la hora
df_clima.loc[df_clima.index[-1], "Hora"] = 24
```

```python
df_clima
```

```python
#Para conectarlos hare un merge, los datos de trafico aun no estan tratados del todo pero los de clima y contaminacion si
df_clima_contaminacion = pd.merge(df_contaminacion, df_clima, on=["Año", "Mes", "Dia", "Hora"], how="left")
```

```python
df_final = pd.DataFrame()
for i in df_contaminacion.Estación.unique():
    df_prueba = pd.merge(df_contaminacion[df_contaminacion.Estación==i], df_clima, how="right", on=["Año","Mes","Dia","Hora"])
    df_prueba["Estación"] = df_prueba["Estación"].apply(lambda x: i)
    df_final = pd.concat([df_final,df_prueba], ignore_index=True)
```
